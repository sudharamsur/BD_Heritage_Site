```python
import os
import pandas as pd
import torch
import torch.nn.functional as F
import random
import numpy as np
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from transformers import BlipProcessor, BlipForQuestionAnswering
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm
from torch.nn.utils.rnn import pad_sequence
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from Levenshtein import ratio  # For answer similarity scoring

# Configuration
class Config:
    dataset_path = "/kaggle/input/vqadatasets"
    image_dir = os.path.join(dataset_path, "ImageDataset/ImageDataset")
    annotations_file = os.path.join(dataset_path, "/kaggle/input/vqadatasets/qa.csv")
    model_name = "Salesforce/blip-vqa-base"
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    batch_size = 4
    num_epochs = 50
    learning_rate = 5e-5  # Lower learning rate for stability
    num_workers = 3
    save_path = "best_vqa_model.pth"

# Load dataset
df = pd.read_csv(Config.annotations_file)

# Data Augmentation
transform = transforms.Compose([
    transforms.Resize((384, 384)),
    transforms.RandomApply([
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.RandomRotation(degrees=(-10, 10)),
        transforms.RandomHorizontalFlip(p=0.5),
    ], p=0.7),
    transforms.ToTensor(),
])

# Dataset Class
import nlpaug.augmenter.word as naw  # Import nlpaug word augmentation methods

class VQADataset(Dataset):
    def __init__(self, dataframe, image_dir, processor, transform=None, augment_text=False, num_augment=3):
        self.dataframe = dataframe
        self.image_dir = image_dir
        self.processor = processor
        self.transform = transform
        self.augment_text = augment_text
        self.num_augment = num_augment  # Number of augmented samples per original

        # Initialize text augmenters
        self.synonym_aug = naw.SynonymAug(aug_p=0.3)  # Augment 30% of words with synonyms
        self.swap_aug = naw.RandomWordAug(action="swap", aug_p=0.2)  # Randomly swap words
        self.delete_aug = naw.RandomWordAug(action="delete", aug_p=0.2)  # Randomly delete words

    def augment_question(self, question):
        augmented_questions = [question]
        for _ in range(self.num_augment):
            aug_type = random.choice(["synonym", "swap", "delete"])
            if aug_type == "synonym":
                augmented_questions.append(self.synonym_aug.augment(question))
            elif aug_type == "swap":
                augmented_questions.append(self.swap_aug.augment(question))
            else:
                augmented_questions.append(self.delete_aug.augment(question))
        return augmented_questions

    def __len__(self):
        if self.augment_text:
            return len(self.dataframe) * (self.num_augment + 1)  # Original + augmented
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx % len(self.dataframe)]
        image_id = row["Image ID"]
        question = row["Bangla Questions"]
        answer = row["Bangla Answers"]

        image_path = os.path.join(self.image_dir, f"{image_id}.jpg")
        image = Image.open(image_path).convert("RGB")
        if self.transform:
            image = self.transform(image)

        # Augment text if enabled
        if self.augment_text:
            augmented_questions = self.augment_question(question)
            question = augmented_questions[idx // len(self.dataframe)]

        inputs = self.processor(images=image, text=question, return_tensors="pt", padding=True, do_rescale=False)
        inputs = {k: v.squeeze(0) for k, v in inputs.items()}
        return inputs, answer


# Custom Collate Function
# def collate_fn(batch):
#     inputs = {key: [item[0][key] for item in batch] for key in batch[0][0].keys()}
#     answers = [str(item[1]) if pd.notna(item[1]) else "" for item in batch]

#     for key in inputs:
#         if isinstance(inputs[key][0], torch.Tensor):
#             inputs[key] = pad_sequence(inputs[key], batch_first=True, padding_value=0)

#     return inputs, answers
# Custom Collate Function
def collate_fn(batch, extra_padding=10):  # Add 10 extra padding tokens
    inputs = {key: [item[0][key] for item in batch] for key in batch[0][0].keys()}
    answers = [str(item[1]) if pd.notna(item[1]) else "" for item in batch]

    for key in inputs:
        if isinstance(inputs[key][0], torch.Tensor):
            max_len = max(t.shape[-1] for t in inputs[key]) + extra_padding  # Add extra padding
            padded_tensors = [torch.nn.functional.pad(t, (0, max(0, max_len - t.shape[-1]))) for t in inputs[key]]
            inputs[key] = pad_sequence(padded_tensors, batch_first=True, padding_value=0)

    return inputs, answers



# Load Processor and Model
processor = BlipProcessor.from_pretrained(Config.model_name)
model = BlipForQuestionAnswering.from_pretrained(Config.model_name).to(Config.device)

# Dataset and DataLoader
dataset = VQADataset(df, Config.image_dir, processor, transform)
dataloader = DataLoader(dataset, batch_size=Config.batch_size, shuffle=True, collate_fn=collate_fn, num_workers=Config.num_workers)

# Training Function
# Training Function with Accuracy Calculation
def train(model, dataloader, epochs):
    model.train()
    optimizer = torch.optim.AdamW(model.parameters(), lr=Config.learning_rate)
    best_loss = float("inf")

    train_losses = []   # Store loss per epoch
    train_accuracies = []  # Store accuracy per epoch

    for epoch in range(epochs):
        total_loss, correct_preds, total_samples = 0, 0, 0

        for batch in tqdm(dataloader, desc=f"Epoch {epoch+1}/{epochs}"):
            inputs, answers = batch
            inputs = {k: v.to(Config.device) for k, v in inputs.items()}
            
            # Tokenize answers
            encoded_answers = processor.tokenizer(
                answers, return_tensors="pt", padding=True, truncation=True, max_length=50
            )

            # Forward pass
            outputs = model(**inputs, labels=encoded_answers.input_ids.to(Config.device))
            loss = outputs.loss

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

            # Generate predictions
            preds = model.generate(**inputs)
            decoded_preds = processor.batch_decode(preds, skip_special_tokens=True)

            # Compute accuracy
            correct_preds += sum(p.strip().lower() == a.strip().lower() for p, a in zip(decoded_preds, answers))
            total_samples += len(answers)

        avg_loss = total_loss / len(dataloader)
        accuracy = correct_preds / total_samples

        train_losses.append(avg_loss)  # Store loss
        train_accuracies.append(accuracy)  # Store accuracy

        print(f"Epoch {epoch+1} - Loss: {avg_loss:.4f}, Accuracy: {accuracy:.4f}")

        # Save best model
        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save(model.state_dict(), Config.save_path)
            print(f"Saved Best Model with Loss: {best_loss:.4f}")

    return train_losses, train_accuracies  # Return loss and accuracy lists




# Load Best Model
def load_best_model(model_path):
    model.load_state_dict(torch.load(model_path, map_location=Config.device))
    model.to(Config.device)
    print("Loaded best model for evaluation!")




# Evaluation Function
def evaluate(model, dataloader):
    model.eval()
    all_preds, all_answers = [], []

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            inputs, answers = batch
            inputs = {k: v.to(Config.device) for k, v in inputs.items()}
            outputs = model.generate(**inputs)
            predictions = processor.batch_decode(outputs, skip_special_tokens=True)

            all_preds.extend([pred.strip().lower() for pred in predictions])
            all_answers.extend([str(ans).strip().lower() for ans in answers])

    accuracy = accuracy_score(all_answers, all_preds)
    precision = precision_score(all_answers, all_preds, average='macro', zero_division=0)
    recall = recall_score(all_answers, all_preds, average='macro', zero_division=0)
    f1 = f1_score(all_answers, all_preds, average='macro', zero_division=0)

    return all_answers, all_preds, accuracy, precision, recall, f1

# Plot Metrics
def plot_metrics(losses, accuracies):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(losses) + 1), losses, marker='o')
    plt.title("Training Loss Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(accuracies) + 1), accuracies, marker='o')
    plt.title("Accuracy Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.grid()

    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    train_losses, train_accuracies = train(model, dataloader, Config.num_epochs)
    all_answers, all_preds, accuracy, precision, recall, f1 = evaluate(model, dataloader)
    plot_metrics(train_losses, train_accuracies)

    print("\n--- Final VQA Model Performance ---")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
```
